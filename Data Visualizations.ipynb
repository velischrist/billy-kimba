```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import git
repo = git.Repo("./", search_parent_directories=True)
homedir = repo.working_dir
```

```python
aggdf = pd.read_csv(f"{homedir}/data/international/covid/our_world_in_data/full_data.csv")
aggdf.head()
```

```python
aggdf = aggdf[(aggdf['location'] == 'United States') & (aggdf['total_deaths'] > 0)]
```

```python
aggdf.reset_index(inplace=True, drop=True)
```

```python
aggdf.head()
```

```python
aggdf.tail()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(aggdf['total_cases'], aggdf['total_deaths'], s=7, color='purple')
plt.ticklabel_format(axis="both", style="sci", scilimits=(0,0))
plt.xlabel("Total Cases in U.S.", fontsize=16)
plt.ylabel("Total Deaths in U.S.", fontsize=16)
plt.title("Total Deaths versus Total Cases in the U.S.", fontsize=18)
plt.show()
'''Slight acceleration at the start is likely due to old folks and those at a higher risk initially not
being exposed, but since many senior citizens live in congregated communities/homes, once they were exposed,
we saw this increase in the death rate. However, as susceptible people became much more cautious, we see
the death rate approach and maintain a constant rate.'''
```

```python
plt.figure(figsize=(8,6))
plt.scatter(aggdf.index, aggdf['total_deaths'], s=7, color='red')
plt.xlabel('Days since Date of First Recorded Death', fontsize=16)
plt.ylabel('Total Deaths in U.S.', fontsize=16)
plt.title("Total Deaths in the U.S. versus Time", fontsize=18)
plt.show()
'''The effects of social/physical distancing are shown as the rate of change of total deaths has diverged from 
an exponential path and shows signs of approaching a minimal rate of increase. However, with recent large public
congregations for protesting and some people's non-negligible ignorance of required safety measures that come with
state reopenings will likely lead to a resumed increasing-rate trend in the U.S.'s future aggregate death total. 
However, it will likely not be as extreme as the initial death rate's acceleration since those susceptible are
now informed and thus less likely to expose themselves.'''
```

```python
plt.figure(figsize=(8,6))
plt.scatter(aggdf.index, aggdf['new_deaths'], s=7, color='red')
plt.xlabel('Days since Date of First Recorded Death', fontsize=16)
plt.ylabel('New Daily Deaths in U.S.', fontsize=16)
plt.title('Daily Death Rate versus Time', fontsize=18)
plt.show()
'''This particular graphic implies that the variance in the death rate gradually increased after national social 
distancing policies were implemented on March 16 (roughly 15 days after the date of the first death). It is likely 
a possibility that lack of serious consideration of the virus and its impacts lead to an extended time period 
for citizens to strictly abide by the provided guidelines, as we see that the variance only significantly increases
and the death rate only declines several weeks after March 16. An influx in the number of counties reporting may
also contribute to an initial rapid increase in death rate.'''
```

```python
plt.figure(figsize=(8,6))
plt.scatter(aggdf.index, aggdf['new_cases'], s=7, color='magenta')
plt.xlabel('Days since Date of First Recorded Death', fontsize=16)
plt.ylabel('New Daily Cases in U.S.', fontsize=16)
plt.title('Daily Case Rate versus Time', fontsize=18)
plt.show()
'''We can see similar results in the respective daily U.S. case counts. However, notice the offset in time between 
the acceleration of new cases rate and new deaths rate - the fact that the case rate spikes a few weeks earlier 
than the death rate provides confirmation of the time delay between point of infection and death.'''
```

```python
df = pd.read_csv(f"{homedir}/PreliminaryModels/completed_master_X.csv")
df.tail()
```

```python
import math
j = list(df['mobility'])
count = 0
for m in j:
    if math.isnan(m):
        count += 1
print(count)
print(len(j))
```

```python
snohofips = 53061
snho = df[df['fips'] == snohofips].reset_index(drop=True)
mob = list(snho['mobility'])
deaths = list(snho['deaths'])
cases = list(snho['cases'])
mobb = []
deathss = []
casess = []
for i in range(len(mob)):
    if not math.isnan(mob[i]) and mob[i] != 0:
        mobb.append(mob[i])
        deathss.append(deaths[i])
        casess.append(cases[i])
mobb = list(np.array(mobb) / max(mobb))
deathss = list(np.array(deathss) / max(deathss))
casess = list(np.array(casess) / max(casess))
```

```python
plt.figure(figsize=(10,6))
plt.plot(list(range(len(deathss))), deathss, linewidth=0.5, color='red')
plt.plot(list(range(len(deathss))), casess, linewidth=0.5, color='magenta')
plt.plot(list(range(len(deathss))), mobb, linewidth=0.5, color='blue')
plt.xlabel("Days since Date of First Death", fontsize=16)
plt.ylabel("Death, Case Rates and Mobility (Normalized)", fontsize=16)
plt.title("Daily Death and Case Rates versus Mobility for Snohomish County", fontsize=18)
plt.legend(['Daily Death Rate', 'Daily Case Rate', 'Daily Mobility'])
plt.show()
'''First we'll take a look at a couple different counties' and how mobility impacts daily case and death rates. 
Snohomish County, Washington is the first county we'll look at of the two. Snohomish was one of the earliest hit counties and 
was one of the more consistent counties in reporting from the start. One can clearly deduce from the peaks of 
each of the lines that mobility is correlated with daily case and death rates, but their correlation is slightly 
delayed for case rates, and even more delayed for death rates. When mobility spikes, the corresponding spike in 
case rate occurs just over a week after, and the corresponding spike in death rate occurs over 2 weeks after. 
This lends to why mobility data is so valuable to incorporate in predictive models related to COVID-19.'''
```

```python
lafips = 36103
la = df[df['fips'] == lafips].reset_index(drop=True)
mob = list(la['mobility'])
deaths = list(la['deaths'])
cases = list(la['cases'])
mobb = []
deathss = []
casess = []
for i in range(len(mob)):
    if not math.isnan(mob[i]) and mob[i] != 0:
        mobb.append(mob[i])
        deathss.append(deaths[i])
        casess.append(cases[i])
mobb = list(np.array(mobb) / max(mobb))
deathss = list(np.array(deathss) / max(deathss))
casess = list(np.array(casess) / max(casess))


plt.figure(figsize=(10,6))
plt.plot(list(range(len(deathss))), deathss, linewidth=0.5, color='red')
plt.plot(list(range(len(deathss))), casess, linewidth=0.5, color='magenta')
plt.plot(list(range(len(deathss))), mobb, linewidth=0.5, color='blue')
plt.xlabel("Days since Date of First Death", fontsize=16)
plt.ylabel("Death, Case Rates and Mobility (Normalized)", fontsize=16)
plt.title("Daily Death and Case Rates versus Mobility for Suffolk County, NY", fontsize=18)
plt.legend(['Daily Death Rate', 'Daily Case Rate', 'Daily Mobility'])
plt.show()
'''For the second county, we'll look at Suffolk County, New York. Suffolk county was impacted on much more 
damaging scale than Snohomish, but there are noticeable commonalities in the time delays between spikes in 
mobility and spikes in daily case and death rates. For each peak in mobility, the corresponding peak in case 
rate appears to be roughly 2 weeks after, and the corresponding peak in death rate appears to be over a
full month after. These extended delays relative to Snohomish may be due to delays in reporting from the 
overwhelming nature of coronavirus's impact on densely populated New York counties compared to typical counties.
These results could also come from flaws in Suffolk county's system of figure reporting.'''
```
